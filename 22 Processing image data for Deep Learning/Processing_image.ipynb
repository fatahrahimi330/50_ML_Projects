# Image data processing

## Reading an image file

```python
# importing the image module from matplotlib library
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
```

## converting to a numpy array

```python
# loading an image through matplotlib.image module
img = mpimg.imread('cat.jpg')
img
```

```python
img.shape
```

```python
# displaying the image from numpy array
plt.imshow(img)
```

## Resizing an image

```python
from PIL import Image 
```

```python
img = Image.open('cat.jpg')
img_resized = img.resize((200, 200))
img_resized.save('cat_image_resized.jpg')
```

```python
img_res = mpimg.imread('cat_image_resized.jpg')
plt.imshow(img_res)
```

```python
img_res.shape
```

## RGB to Grayscale conversion

```python
import cv2
import os

print(os.getcwd())  # See where Python is looking for cat.jpg
img = cv2.imread('cat_image_resized.jpg')

if img is None:
    print("⚠️ Image not found. Please check the path!")
else:
    print("✅ Image loaded successfully!")
```

```python
# importing OpenCV library
import cv2
img = cv2.imread('cat_image_resized.jpg')
grayscale_image = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```python
import matplotlib.pyplot as plt
plt.imshow(grayscale_image, cmap='gray')
plt.axis('off')  # Hide axis numbers
plt.title('Grayscale Image')
plt.show()
```
